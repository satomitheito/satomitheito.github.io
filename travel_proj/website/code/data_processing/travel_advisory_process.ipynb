```python
import pandas as pd
```

```python
travel_advisories = pd.read_csv('../../data/raw/Travel_Advisories.csv')
```

```python
travel_advisories.dtypes
```

```python
travel_advisories
```

```python
# Change column type 
travel_advisories["Country"] = travel_advisories["Country"].astype(str)
travel_advisories["Advisory Level"] = travel_advisories["Advisory Level"].astype(str)
travel_advisories["Date Updated"] = pd.to_datetime(travel_advisories["Date Updated"])
```

```python
travel_advisories
```

```python
# Have column for level number since advisory level is in str
travel_advisories["Advisory Level Number"] = travel_advisories["Advisory Level"].str.extract(r'Level (\d)').astype(float)
```

```python
travel_advisories.to_csv("../../data/processed/Travel_Advisories_Processed.csv", index=False)
```
